```python
from melo.api import TTS
from IPython.display import Audio
import gradio as gr
```

```python
# Initialize MeloTTS TTS model
device = 'auto' # Automatically uses GPU if available

# Function to generate speech and return the output
def generate_speech(text):
    language_code = "EN"
    speed = 1

    model = TTS(language=language_code, device=device)

    speaker_ids = model.hps.data.spk2id
    selected_speaker_id = speaker_ids["EN-Default"]

    output_path = 'generated_MeloTTS.wav'

    # Generate speech and save to file
    model.tts_to_file(text, selected_speaker_id, output_path, speed=speed)

    return output_path

```

```python
# Create Gradio Interface
def gradio_interface():
    interface = gr.Interface(
        fn=generate_speech,
        inputs=[
            gr.Textbox(label="Enter Text", lines=10)
        ],
        outputs=gr.Audio(label="Generated Speech"),
        title="MeloTTS Text-to-Speech"
    )

    interface.launch()

# Launch the Gradio interface
gradio_interface()
```
